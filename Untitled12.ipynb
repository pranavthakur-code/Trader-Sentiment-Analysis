```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

plt.style.use('seaborn-v0_8')
```

```python
print("Current Directory:", os.getcwd())
print("Files Available:", os.listdir())
```

```python
trades = pd.read_csv("historical_data.csv")
sentiment = pd.read_csv("fear_greed_index.csv")

print("Trades Shape:", trades.shape)
print("Sentiment Shape:", sentiment.shape)
```

```python
print(trades.head())
print(sentiment.head())

print("\nMissing Values (Trades):")
print(trades.isnull().sum())

print("\nMissing Values (Sentiment):")
print(sentiment.isnull().sum())

print("\nDuplicate Trades:", trades.duplicated().sum())
print("Duplicate Sentiment:", sentiment.duplicated().sum())
```

```python
# Convert trade timestamp
trades['Timestamp IST'] = pd.to_datetime(trades['Timestamp IST'], errors='coerce')

# Create daily date column
trades['date'] = trades['Timestamp IST'].dt.date

# Convert sentiment date
sentiment['date'] = pd.to_datetime(sentiment['date'], errors='coerce')
sentiment['date'] = sentiment['date'].dt.date
```

```python
data = trades.merge(
    sentiment[['date', 'classification']],
    on='date',
    how='left'
)

print("Merged Shape:", data.shape)
data.head()
```

```python
daily_pnl = data.groupby(['Account','date'])['Closed PnL'].sum().reset_index()
daily_pnl.head()
```

```python
data['win'] = np.where(data['Closed PnL'] > 0, 1, 0)

win_rate = data.groupby('Account')['win'].mean().reset_index()
win_rate.head()
```

```python
trades_per_day = data.groupby('date').size()

plt.figure(figsize=(10,5))
trades_per_day.plot()
plt.title("Trades Per Day")
plt.show()
```

```python
long_short = data.groupby(['classification','Side']).size().unstack()

long_short.plot(kind='bar', figsize=(8,5))
plt.title("Long vs Short by Sentiment")
plt.show()
```

```python
performance = data.groupby('classification')['Closed PnL'].mean()
print(performance)

sns.boxplot(x='classification', y='Closed PnL', data=data)
plt.title("PnL Distribution: Fear vs Greed")
plt.show()
```

```python
freq = data.groupby('classification').size()

freq.plot(kind='bar')
plt.title("Trade Frequency by Sentiment")
plt.show()
```

```python
# Re-process sentiment data correctly within this cell to ensure 'classification' is available.
# This addresses the errors that occurred in cells 8Aif8nHwwlE- (date conversion) and s-2Q-VBYwnU1 (column name in merge) indirectly.
sentiment_corrected = sentiment.copy()
sentiment_corrected['date'] = pd.to_datetime(sentiment_corrected['date'], errors='coerce').dt.date

# Ensure trades 'date' is also correct for merging (though it was likely correct).
trades_corrected = trades.copy()
trades_corrected['Timestamp IST'] = pd.to_datetime(trades_corrected['Timestamp IST'], errors='coerce')
trades_corrected['date'] = trades_corrected['Timestamp IST'].dt.date

# Re-merge the data with the correct classification column.
data_transformed = trades_corrected.merge(
    sentiment_corrected[['date', 'classification']], # Corrected 'Classification' to 'classification'
    on='date',
    how='left'
)

# Now use data_transformed for the plot
sns.boxplot(x='classification', y='Size USD', data=data_transformed) # Use 'classification' here too
plt.title("Trade Size by Sentiment")
plt.show()
```

```python
median_size = data['Size USD'].median()

data['Size_Segment'] = np.where(
    data['Size USD'] > median_size,
    'High Size',
    'Low Size'
)

segment_perf = data.groupby(['Size_Segment','classification'])['Closed PnL'].mean()

print(segment_perf)
```
